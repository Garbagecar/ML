# <font size=10>聚类算法<font>
<font size=4>   
    
**什么是聚类算法**
- 聚类是按照某个特定标准(如距离、密度)把一个数据集分割成不同的类或簇
- 同一个簇内的数据对象的相似性尽可能大，同时不在同一个簇中的数据对象的差异性也尽可能地大
- 也即聚类后同一类的数据尽可能聚集到一起，不同类数据尽量分离
    
**聚类与分类**
- 聚类：是指把相似的数据划分到一起，具体划分的时候**并不关心这一类的标签**，目标就是把相似的数据聚合到一起，聚类是一种**无监督**学习方法
- 分类：是把不同的数据划分开，其过程是通过训练数据集获得一个分类器，再**通过分类器去预测未知数据**，分类是一种**有监督**学习方法
    
**聚类算法的一般流程**
1. 数据准备：**特征标准化**和**降维**
2. 特征选择：从最初的特征中选择最有效的特征，并将其存储在向量中
3. 特征提取：通过对选择的特征进行转换形成新的突出特征
4. 聚类：基于某种距离函数进行相似度度量，获取簇
5. 聚类结果评估：分析聚类结果，比如距离误差和(SSE)等
    
**聚类算法的分类**
- 划分式聚类方法：KMeans、Kmeans++、MiniBatch KMeans
- 基于密度的聚类方法：DBSCAN
    

## <font size=6><mark style=background-color:pink>聚类算法一：KMeans<mark><font>

### <font size=6>一、KMeans算法概述<font>
<font size=4>
    
**算法引例**：牧师-村名模型
- 有四个牧师去郊区布道，一开始牧师们随意选了几个布道点，并且把这几个布道点的情况公告给了郊区所有的村民，于是每个村民到离自己家最近的布道点去听课
- 听课之后，大家觉得距离太远了，于是每个牧师统计了一下自己的课上所有的村民的地址，搬到了所有地址的中心地带，并且在海报上更新了自己的布道点的位置
- 牧师每一次移动不可能离所有人都更近，有的人发现A牧师移动以后自己还不如去B牧师处听课更近，于是每个村民又去了离自己最近的布道点……
- 就这样，牧师每个礼拜更新自己的位置，村民根据自己的情况选择布道点，最终稳定了下来   
- **牧师的目的是为了让每个村民到其最近中心点的距离和最小**
    
**算法步骤**
- 选择初始化的K个样本作为初始聚类中心：$a_{1}、a_{2}、a_{3}...a_{K}$
- 针对数据集中每个样本 $x_{i}$ 计算它到K个聚类中心的距离，将其分到距离最小的聚类中心所对应的类中
- 针对每个类别$c_{j}$，重新计算它的聚类中心$a_{j}=\frac{1}{|c_{j}|}\sum_{x∈c_{j}}x$，即属于该类的所有样本的质心
- 重复前两步操作，直到达到某个中止条件（迭代次数达到一定值、误差的变化降到一定阈值下等）
    
**特征工程**
- 机器学习中有部分模型是基于距离度量进行模型预测和分类的，由于距离对特征之间不同取值范围非常敏感，所以基于距离度量的模型是十分有必要做**数据标准化**处理的
- 最典型基于距离度量的模型包括**K近邻、kmeans聚类、感知机和SVM**
- 线性回归类的几个模型一般情况下也是需要做数据标准化处理的
- 逻辑回归、决策树、基于决策树的Boosting和Bagging等集成学习模型对于特征取值大小并不敏感，一般不需要做数据标准化处理
- 另外有较多类别变量的数据也是不需要做标准化处理的

**算法评估**
1. **误差平方和Sum of the Squared Errors**：$SSE=\sum_{j=1}^{K}\sum_{i=1}^{|c_{j}|}(x_{i}-a_{j})^{2}$
2. **轮廓系数Silhouette Coefficient**：$s=\frac{b-a}{max\{a,b\}}$
- a表示样本点与同一簇中所有其他点的距离求和后再平均(平均距离)，即样本点与同一簇中其他点的相似度
- 再求样本到每个不含自己的簇中所有点的平均距离，再找到平均距离的最小值b，即样本点与下一个最近簇中其他点的相似度
- K-Means追求的是对于每个簇而言，其簇内差异小，而簇外差异大，轮廓系数S正是描述簇内外差异的关键指标
- 由公式可知，S取值范围为(-1, 1)，**当S越接近于1，则聚类效果越好，越接近-1，聚类效果越差**
- **对于簇结构为凸的数据轮廓系数较高，对于簇结构非凸的轮廓系数较低**
- **因此，轮廓系数不能在不同的算法之间比较优劣，如统一数据下，可能KMeans的结果就比DBSCAN要好(最好不要用在DBSCAN中)**
    
**优点**：
- 原理比较简单，实现也是很容易，收敛速度快
- 聚类效果较优
- 算法的可解释度比较强
- 主要需要调参的参数仅仅是簇数k。

**缺点**：
- K值的选取不好把握(改进：肘部法则)
- 对于不是凸的数据集比较难收敛(改进：基于密度的聚类算法更加适合，比如DESCAN算法)
- 如果数据的类型不平衡，比如数据量严重失衡或者类别的方差不同，则聚类效果不佳
- 采用迭代方法，得到的结果只是局部最优，这和误差平方和是非凸函数有关
- 对噪音和异常点比较的敏感(改进1：离群点检测的LOF算法，通过去除离群点后再聚类，可以减少离群点和孤立点对于聚类效果的影响；改进2：改成求点的中位数，这种聚类方式即K-Mediods聚类（K中值））
- 初始聚类中心的选择影响聚类效果(改进1：k-means++;改进2：二分K-means)

**优化方法**
    
1. **K值的选取：肘部法则**
- 画出不同K值的SSE走势图，在SSE走势图下降最快的“肘部”对应的K值作为最优


    
- 像在上图中，选择K=3为最优
    
2. **初始聚类中心的选择：KMeans++**
- 按照均匀分布(每个样本被选取的概率是一样的)，从点集X中选择首次聚类的第一个聚类中心$a_{1}$
- 按照概率$ \frac{d(x)^2}{\sum_{x \in  \chi }d(x)^2}$，选则第二个聚类中心$a_{2}$，d(x)表示x到第一个聚类中心$a_{1}$的距离，即离第一个聚类中心越远的点被选中的概率越大  
- 重复上一步，直到选完指定的K个中心  

3. **减少计算量：elkan算法**
    


- 根据三角形两边之和大于第三边: $c+d>a$  
- 如果$c<\frac{a}{2}$ 即 $a>2c$，那么$d>c$，所以样本1属于聚类中心1所代表的簇，就减少了一次计算距离d的开销
<font>
    
4. **MiniBatch KMeans**
- Mini Batch K-Means算法是K-Means算法的变种，采用小批量的数据子集减小计算时间
- 这里所谓批是指每次训练算法时所随机抽取的数据子集，采用这些随机产生的子集进行训练算法，大大减小了计算时间，结果一般只略差于标准算法。
- 算法的迭代步骤有两步：①从数据集中随机抽取一些数据形成小批量，把他们分配给最近的质心 ②根据重新分配给这个簇和这个簇原先已有的样本来更新质心  
- 与K均值算法相比，数据的更新是在每一个小样本集上
- Mini Batch K-Means比K-Means有更快的收敛速度，略微降低聚类的效果


### <font size=6>二、sklearn中的KMeans算法<font>
<font size=4>

**KMeans**
- 类：**sklearn.cluster.KMeans**
- 实例化语法：**sklearn.cluster.KMeans(n_clusters=,init='k-means++'..)**
- n_clusters: 即我们的k值，一般需要多试一些值以获得较好的聚类效果
- max_iter： 最大的迭代次数，一般如果是凸数据集的话可以不管这个值，如果数据集不是凸的，可能很难收敛，此时可以指定最大的迭代次数让算法可以及时退出循环
- n_init：用不同的初始化质心运行算法的次数，由于K-Means是结果受初始值影响的局部最优的迭代算法，因此需要多跑几次以选择一个较好的聚类效果，默认是10，一般不需要改，如果你的k值较大，则可以适当增大这个值
- init： 即初始值选择的方式，可以为完全随机选择'random'，优化过的'k-means++'或者自己指定初始化的k个质心，一般建议使用默认的'k-means++'
- algorithm：有“auto”, “full” or “elkan”三种选择。"full"就是我们传统的K-Means算法， “elkan”是我们原理篇讲的elkan K-Means算法。默认的"auto"则会根据数据值是否是稀疏的，来决定如何选择"full"和“elkan”。一般数据是稠密的，那么就是 “elkan”，否则就是"full"。一般来说建议直接用默认的"auto"
- random_state

具体应用：
- from sklearn.cluster import KMeans
- km = KMeans(n_clusters=4)
- km.fit(X)
- X_predict = km.predict(X)
    
**MiniBatch KMeans**
- 类：**sklearn.cluster.MiniBatchKMeans**
- 实例化语法：**sklearn.cluster.KMeans(n_clusters=,init='k-means++'..)**
- n_clusters: 即k值，和KMeans类的n_clusters意义一样。
- max_iter：最大的迭代次数， 和KMeans类的max_iter意义一样。
- n_init：用不同的初始化质心运行算法的次数。这里和KMeans类意义稍有不同，KMeans类里的n_init是用同样的训练集数据来跑不同的初始化质心从而运行算法。而MiniBatchKMeans类的n_init则是每次用不一样的采样数据集来跑不同的初始化质心运行算法。
- batch_size：即用来跑Mini Batch KMeans算法的采样集的大小，默认是100.如果发现数据集的类别较多或者噪音点较多，需要增加这个值以达到较好的聚类效果。
- init： 即初始值选择的方式，和KMeans类的init意义一样。
- init_size: 用来做质心初始值候选的样本个数，默认是batch_size的3倍，一般用默认值就可以了。
- reassignment_ratio: 某个类别质心被重新赋值的最大次数比例，这个和max_iter一样是为了控制算法运行时间的。这个比例是占样本总数的比例，乘以样本总数就得到了每个类别质心可以重新赋值的次数。如果取值较高的话算法收敛时间可能会增加，尤其是那些暂时拥有样本数较少的质心。默认是0.01。如果数据量不是超大的话，比如1w以下，建议使用默认值。如果数据量超过1w，类别又比较多，可能需要适当减少这个比例值。具体要根据训练集来决定。
- max_no_improvement：即连续多少个Mini Batch没有改善聚类效果的话，就停止算法， 和reassignment_ratio， max_iter一样是为了控制算法运行时间的。默认是10.一般用默认值就足够了。

#### KMeans

```python
#导入KMeans算法的类
from sklearn.cluster import KMeans 

#导入Matplotlib库看聚类效果
from matplotlib import pyplot as plt

#导入numpy库读取文本数据
import numpy as np
```

```python
#导入数据
data = np.genfromtxt(r"E:\jupyterlab\ML\kmeans.txt",delimiter=' ')

#绘制散点图看数据分布情况
plt.scatter(data[:,0],data[:,1],marker='o') #散点图
plt.show()
```

```python
#看看数据情况,80个数据，两个特征
data.shape
```

```python
# 设置k值
k = 4

#实例化KMeans算法的估计器
km = KMeans(n_clusters=k,init="k-means++",algorithm="elkan",random_state=2022,verbose=1)
#km = KMeans(n_clusters=k,init="k-means++",algorithm="elkan",random_state=2022)

#用数据集训练KMeans算法得到模型
km.fit(data)
```

```python
#查看聚类后的4个质心
km.cluster_centers_
```

```python
#查看每个样本的标签
km.labels_ 
```

```python
#查看聚类迭代了几次
km.n_iter_  
```

```python
#SSE
km.inertia_
```

```python
#预测，执行分类
km.predict(data)
```

```python
#可视化聚类结果

lables = km.labels_
centers = km.cluster_centers_ 

# 画样本点，用颜色表示这4个不同的类别
mark = ['or','ob','og','oy']
for i,d in enumerate(data):
    plt.plot(d[0],d[1],mark[lables[i]])
    
# 画分类中心点
mark = ['*r','*b','*g','*y']
for i,center in enumerate(centers):
    plt.plot(center[0],center[1],mark[i],markersize=20)
plt.show()
```

```python
#根据肘部法则，选择合适的K值

SSE=[]
for k in range(2,30):
    km = KMeans(n_clusters=k)
    km.fit(data)
    SSE.append(km.inertia_)
plt.plot(range(2,30),SSE)
plt.xlabel('k')
plt.ylabel('SSE')
plt.show()
```

## <font size=6><mark style=background-color:pink>聚类算法二：DBSCAN<mark><font>

### <font size=6>一、DBSCAN算法概述<font>
<font size=4>
    
**算法背景**
- DBSCAN(Density-Based Spatial Clustering of Applications with Noise)是⼀种基于密度的聚类⽅法，旨在寻找被低密度区域分离的⾼密度区域
- DBSCAN是⼀种简单、有效的基于密度的聚类算法，且包含了基于密度的许多⽅法中的重要概念

**核心概念**
- $\epsilon$邻域：给定对象半径为$\epsilon$的区域称为该对象的$\epsilon$邻域
- 核心对象：如果给定对象$\epsilon$邻域内的样本点数大于等于阈值$t$，则称该对象为核心对象
- 直接密度可达：对于样本集合D，如果样本点Q在P的$\epsilon$邻域内，并且P为核心对象，那么P到Q直接密度可达
- 密度可达：对于样本集合D，给定一串样本点P1、P2....Pn，P=p1,Q=pn,假如对象Pi-1到Pi直接密度可达，那么对象P到Q密度可达
- 密度相连：存在样本集合D中的一点O，如果对象O到对象P和对象Q都是密度可达的，那么P和Q密度相联
    
**算法步骤**
- 指定合适的$\epsilon$和阈值$t$ 
- 计算所有的样本点，如果p的$\epsilon$邻域里有超过$t$个点，则创建一个以p为核心点的新簇
- 反复寻找这些核心点直接密度可达,密度可达的点，将其加入到相应簇，对于核心点发生“密度相连”状况的簇，给予合并
- 当没有新的点可以被添加到任何簇时，算法结束

**优点**：
- 与K-means方法相比，DBSCAN不需要事先知道要形成的簇类的数量。
- 与K-means方法相比，DBSCAN可以发现任意形状的簇类
- 同时，DBSCAN能够识别出噪声点
- DBSCAN对于数据库中样本的顺序不敏感，即Pattern的输入顺序对结果的影响不大
- 但是，对于处于簇类之间边界样本，可能会根据哪个簇类优先被探测到而其归属有所摆动

**缺点**：
DBSCAN使用了统一的邻域半径$\epsilon$和邻域中样本的阈值，在类中的数据分布密度不均匀的情况，$\epsilon$较小时，密度小的cluster会被划分成多个性质相似的cluster，$\epsilon$较大时，会使得距离较近且密度较大的cluster被合并成一个cluster，所以在高维数据时，因为维数灾难问题，$\epsilon$的选取比较困难
- DBScan不能很好反映高维数据
- DBScan不能很好反映数据集以变化的密度。
- 如果样本集的密度不均匀、聚类间距差相差很大时，聚类质量较差



### <font size=6>二、sklearn中的DBSCAN算法<font>
<font size=4>

**KMeans**
- 类：**sklearn.cluster.DBSCAN**
- 实例化语法：**sklearn.cluster.dbscan(..)**
- eps：从一个观察值到另一个观察值得最远距离，超过这个距离将不再认为二者是邻居
- min_samples：最小限度的邻居数量，如果一个观察值在其周围小于eps距离的范围内有超过这个数量的邻居，就被认为是核心观察值
- metric：eps所用的距离度量，比如minkowski(闵可夫斯基距离)或者euclidean(欧氏距离)。注意，如果使用闵可夫斯基距离，就可以用参数p设定闵可夫斯基中的幂次
- algorithm：{'auto', 'ball_tree', 'kd_tree', 'brute'}，可选
- p：闵可夫斯基距离公式中的p
- leaf_size：使用'ball_tree'或'kd_tree'时停止建子树的叶子节点数量的阈值

### <font size=6>三、DBSCAN算法案例<font>

```python
#导入numpy用于拼接样本
import numpy as np

#导入matplotlib可视化聚类结果
import matplotlib.pyplot as plt

#导入sklean中的datasets模块，自建聚类样本集
from sklearn import datasets
```

```python
#自建两个不同形状的样本集，一个样本集有2000个样本，另一个样本集有1000个样本，此时创建的样本带有标签
#datasets.make_circles双环状样本，factor影响两环之间的距离
x1, y1 = datasets.make_circles(n_samples=2000, factor=.5, noise=.05)#x1为坐标，y1为标签，这里不需要标签
#datasets.make_blobs创建正态分布的点集，cluster_std为样本集的标准差(控制离散程度)
x2, y2 = datasets.make_blobs(n_samples=1000, centers=[[1.2,1.2]], cluster_std=[[.1]])

#拼接没有标签的样本集所谓DBSCAN算法用来聚类的样本集
x = np.concatenate((x1, x2))
```

```python
#绘制样本集的分布图
plt.scatter(x[:, 0], x[:, 1], marker='o')
plt.show()
```

```python
#先看一下用KMeans算法聚类的效果

#导入KMeans算法的类
from sklearn.cluster import KMeans

#实例化KMeans的估计器，并带入样本集进行训练
y_pred = KMeans(n_clusters=3).fit_predict(x)

#查看KMeans聚类结果
plt.scatter(x[:, 0], x[:, 1], c=y_pred)
plt.show()
```

```python
#现在使用DBSCAN算法进行聚类

#导入DBSCAN算法的类
from sklearn.cluster import DBSCAN

#实例化DBSCAN的估计器，并带入样本集进行训练
y_pred = DBSCAN(eps = 0.2,min_samples=50).fit_predict(x)

#查看DBSCAN聚类结果
plt.scatter(x[:, 0], x[:, 1], c=y_pred)
plt.show()
```
